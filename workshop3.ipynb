```python
import cv2
import matplotlib.pyplot as plt
```

```python
original_image=cv2.imread('TEDDY.jpg')
```

```python
plt.imshow(original_image)
```

```python
blur_image=cv2.blur(original_image,ksize=(3,3))
plt.imshow(blur_image)
```

```python
import numpy as np
medium=np.median(original_image)
```

```python
low=int(max(0,0.7
            *medium))
high=int(max(255,1/3*medium))
```

```python
canny_image=cv2.Canny(blur_image,threshold1=low,threshold2=high)
```

```python
plt.imshow(canny_image)
```
